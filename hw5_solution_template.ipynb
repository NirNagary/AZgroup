```python
data.pivot_table('AveragePrice', index =['region','year'], columns = 'type_organic')
```

```python
data['region']=data['region'].astype('category').cat.codes
features = ['region', 'type_organic', 'AveragePrice', 'Total Bags']
X = data[features]
y = data['total_sold']
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
dt = sk.tree.DecisionTreeRegressor(random_state=0)
dt.fit(X_train, y_train)


def mse(a,b): #function to RMSE
    return np.sqrt((np.square(a-b)).mean())
print("The predictions are:")
test_pred = dt.predict(X_test)
print(test_pred)
print("RMSE: {:.3f}".format(mse(y_test.values,test_pred)))
print('the mse is very high so the model is not predecting well')
```
